```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import LabelEncoder
```

# Load Dataset

The House Price India dataset is loaded using Pandas. This dataset contains various property-related features that will be enhanced using feature engineering techniques.

```python
df = pd.read_csv("House Price India.csv")

df.head()
```

# Dataset Exploration

Before applying feature engineering, the dataset is explored to understand its structure, dimensions, data types, and statistical properties.

```python
print("Dataset Shape:", df.shape)
```

```python
df.info()
```

```python
df.describe()
```

# Missing Value Handling

Missing values are handled before feature engineering.

- Numerical columns are filled using the median.
- Categorical columns are filled using "Unknown".

```python
df.fillna(df.median(numeric_only=True), inplace=True)

df.fillna("Unknown", inplace=True)

df.isnull().sum()
```

# Feature Engineering

New features are created using existing columns to improve the information available for machine learning models.

```python
df["Total Rooms"] = (
    df["number of bedrooms"] +
    df["number of bathrooms"]
)
```

```python
df["House Age"] = (
    2025 -
    df["Built Year"]
)
```

```python
df["Living Area Ratio"] = (
    df["living area"] /
    df["lot area"]
)
```

```python
df[
    [
        "Total Rooms",
        "House Age",
        "Living Area Ratio"
    ]
].head()
```

# Label Encoding

The Date column is converted into numerical values using Label Encoding.

```python
label = LabelEncoder()

df["Date"] = label.fit_transform(
    df["Date"]
)

df[["Date"]].head()
```

# One-Hot Encoding

The "number of floors" column is converted into binary columns using One-Hot Encoding.

```python
df = pd.get_dummies(
    df,
    columns=["number of floors"],
    drop_first=True
)

df.head()
```

# Log Transformation

The Price column is transformed using the logarithmic function to reduce skewness and improve data distribution.

```python
df["Price"] = np.log1p(
    df["Price"]
)

df["Price"].head()
```

# Enhanced Dataset

The dataset now contains:

- New engineered features
- Encoded categorical variables
- Log-transformed target variable

The processed dataset is ready for machine learning applications.

```python
df.head()
```

# Export Enhanced Dataset

The enhanced dataset is exported as a CSV file for future machine learning tasks.

```python
df.to_csv(
    "Enhanced_House_Price_India.csv",
    index=False
)

print("Enhanced dataset exported successfully.")
```
